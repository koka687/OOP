 - створити гру, карткову гру, в якій кожна кортка це є якийсь ітем типу зброя, гравець витягує одну картку та кладе на стіл, дві карти двоїх гравців змагаються між собою, програє та в якої закінчилась живучість

- це проста програма написана за допомогою функціонального програмування

```python
import random

def calculate_critical_hit(name, attack):
    if random.random() < 0.2:  # 20% chance for a critical hit
        print(f"Critical Hit from {name}!")
        return attack * 2
    return attack

list_of_cards = [
{
    "name": "Sword of Destiny",
    "attack": 5,
    "health": 35,
    "defense": 1,
    },
    {
    "name": "Shield of Valor",
    "attack": 2,
    "health": 50,
    "defense": 2,
    },
    {
    "name": "Bow of Eternity",
    "attack": 4,
    "health": 40,
    "defense": 0,
    },
]

player_one_deck = random.choice(list_of_cards)
player_two_deck = random.choice(list_of_cards)

while player_one_deck["health"] > 0 and player_two_deck["health"] > 0:
    print("Новий хід!")
    print(player_one_deck.keys())

    print(f"{player_one_deck['name']} атакує {player_two_deck['name']}")
    player_two_deck["health"] = player_two_deck["health"] + player_two_deck["defense"] - calculate_critical_hit(player_one_deck["name"], player_one_deck["attack"])
    if player_two_deck["health"] <= 0:
        print(f"{player_one_deck['name']} wins!")
        break
    print(f"{player_two_deck['name']} атакує {player_one_deck['name']}")
    player_one_deck["health"] = player_one_deck["health"] + player_one_deck["defense"] - calculate_critical_hit(player_two_deck["name"], player_two_deck["attack"])
    if player_one_deck["health"] <= 0:
        print(f"{player_two_deck['name']} wins!")
        break
```

- якщо прості методи починаються з _ або __ це є приховані методи
- класові методи або ательнативні конструктори

```python
class Person:
    _object_count = 0  # Класова змінна для підрахунку об'єктів
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        Person._count_objects_created()
    
    @classmethod
    def from_full_name(cls, full_name: str):
        "Повне ім'я у форматі 'Імя Прізвище'"
        name, surname = full_name.split(" ")
        return cls(name, surname)
    
    @classmethod
    def from_dict(cls, info: dict):
        "Ініціалізація з словника {'name': 'Імя', 'surname': 'Прізвище'}"
        return cls(info["name"], info["surname"])
    
    @classmethod
    def _count_objects_created(cls):
        "Підрахунок створених обєктів"
        cls._object_count += 1
        return cls._object_count
```

```python
a = Person("Ivan", "Ivanov")
print(a._object_count)
b = Person.from_full_name("Petr Petrov")
c = Person.from_dict({"name": "Sidor", "surname": "Sidorov"})

print(a.name, a.surname, a._object_count)
print(b.name, b.surname)
print(c.name, c.surname)

name = "Петро Іванович"
name, surname = name.split(" ")
d = Person(name, surname)

print(f"Створено обєктів Person: {Person._object_count}, {d._object_count}")
```
