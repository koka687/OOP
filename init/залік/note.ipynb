Створити клас PairwiseSummer, який:

- Приймає в конструкторі два списки однакової довжини
- Має метод calculate_sums(), який обчислює попарні суми елементів
- Має property sum_list, яка повертає список попарних сум

```python
class PairwiseSummer:
    def __init__(self, list1, list2):
        self._list1 = list1
        self._list2 = list2
        self._result = []

    def calculate_sums(self):
        self._result = [a + b for a, b in zip(self._list1, self._list2)]
        return self._result

    @property
    def sum_list(self):
        if not self._result:
            self.calculate_sums()
        print(f"Результат попарного додавання: {self._result}")
        return self._result

    def __str__(self):
        return "Виконала: Анна, Група: КН-33"

# Демонстрація роботи програми
data_sets = [
    ([1, 2, 3], [4, 5, 6]),
    ([10, 20], [30, 40]),
    ([0.5, 1.5, 2.5], [0.5, 0.5, 0.5])
]

for l1, l2 in data_sets:
    summer = PairwiseSummer(l1, l2)
    print(summer)
    _ = summer.sum_list
```
