```python

from datetime import datetime

class MyAnimals:
    max_age = 20 
    senior_age_threshold = 10 

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        """Розраховує очікуваний рік смерті тварини."""
    
        return datetime.today().year - self.age + self.max_age
    
    @property
    def is_senior(self):
        """Перевіряє, чи вік тварини перевищує пороговий вік."""
        return self.age >= self.senior_age_threshold
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.name} прожила за мною один рік. Тепер їй {self.age} років."

a1 = MyAnimals("Кіт", "Мурка", 5)
a2 = MyAnimals("Собака", "Рекс", 11) 


print("\n" + "="*50)
print("Демонстрація властивостей")
print("="*50)

print(f"Мій {a1.animals} ({a1.name}) очікувано буде зі мною до {a1.live_untill} року")

print(f"Чи є {a1.name} старшим (≥ 10 років)? {a1.is_senior}")
print("-" * 50)
print(f"Чи є {a2.name} старшим (≥ 10 років)? {a2.is_senior}") 


print("\n" + "="*50)
print("Зміна атрибута та перерахунок properties")
print("="*50)

print(a1.add_one_year())
print(f"Новий вік {a1.name}: {a1.age} років.")

print(f"Очікуваний рік життя {a1.name} (перераховано): {a1.live_untill}")
print(f"Чи є {a1.name} старшим (перераховано)? {a1.is_senior}")
```

### Методи Класу

- допишіть метод класу який буде превіряти що атрибут name заданий з великої букви

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

```python

class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:

        valid_name = MyAnimals.validate_name(name) 
        
        self.animals = animals
        self.name = valid_name 
        self.age = int(age) 
        MyAnimals.total_objects += 1

    @classmethod
    def validate_name(cls, name: str) -> str:
        """Перевіряє, чи ім'я починається з великої літери, і виправляє, якщо ні."""
        if not name[0].isupper():
            print(f"Попередження: Ім'я '{name}' було виправлено на '{name.capitalize()}'")
            return name.capitalize()
        return name

    @classmethod
    def from_csv(cls, csv_data: str):
        """Створює об'єкт з рядка CSV, використовуючи validate_name всередині __init__."""
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")

        return cls(animal.strip(), name.strip(), age.strip())

print("\n--- Створення a1 (Мурка) ---")
a1 = MyAnimals.from_csv("Кіт,Мурка,5")

print("\n--- Створення a2 (рекс) ---")
a2 = MyAnimals("Собака", "рекс", 2) 

print("\n--- Створення a3 (гуня) ---")
a3 = MyAnimals.from_csv("Рибка,гуня,1")


print("\n" + "="*50)
print("Результати валідації")
print("="*50)
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Мій {a3.animals}, звати {a3.name}, має {a3.age} років")

print(f"\nЗагальна кількість об'єктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```
