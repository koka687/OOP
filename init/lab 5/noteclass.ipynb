- Починає з створення класу Книга.

```python
from enum import Enum

class BookState(Enum):
    AVAILABLE = "Доступна"
    BORROWED = "Позичена"
    RESERVED = "Зарезервована"
    LOST = "Втрачена"

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published
        self.state = BookState.AVAILABLE

    @property
    def description(self):
        return f"Назва: '{self.title}',\n Автор: {self.author},\n Опубліковано в {self.year_published}."
    
    def __str__(self):
        return f"Книга: '{self.title}'"

# Приклад використання
book_shelf = []
for name in ["Посібник Python", "Вивчення JavaScript", "Основи C++",
             "Розробка на Java", "Програмування на Ruby",
             "Введення в Swift", "Основи Go", "Майстерність PHP"]:
    book = Book(title=name, author="Не вказано", year_published=2025)
    book_shelf.append(book)

for book in book_shelf:
    print(book.description)
```

```python
class Student:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        self.student_id = id(self)  # Унікальний номер студентського квитка
        self.borrowed_books: list[Book] = []

    def borrow_book(self, book: Book):
        if book.state != BookState.AVAILABLE:
            print(f"Книга '{book.title}' недоступна для позики.")
            return
        book.state = BookState.BORROWED
        self.borrowed_books.append(book)
        print(f"{self.first_name} взяв книгу '{book.title}'.")
    
    def return_book(self):
        if not self.borrowed_books:
            print(f"{self.first_name} не має книг для повернення.")
            return
        book = self.borrowed_books.pop()
        book.state = BookState.AVAILABLE
        print(f"{self.first_name} повернув книгу '{book.title}'.")

a = Student("Богдан", "Bobas")

import random

print(f"{a.first_name} хоче взяти 3 випадклві книги з бібліотеки.")
for _ in range(3):
    book: Book = random.choice(book_shelf)
    a.borrow_book(book)

print(f"{a.first_name} взяв {len(a.borrowed_books)} книги:")
for b in a.borrowed_books:
    print(f">>> Взято: {b}")

print("Перевіряємо які книги залишились на полицях:")
for b in book_shelf:
    print(f"{b.title}: {b.state.value}")

print(f"{a.first_name} повертає всі свої книги.")
for _ in range(3):
    a.return_book()

print(f"{a.first_name} тепер має {len(a.borrowed_books)} книги.")
for b in book_shelf:
    print(f"{b.title}: {b.state.value}")
```
